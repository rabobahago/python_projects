### Univariate Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f'x_train = {x_train}')
print(f'y_train = {y_train}')
```

#### The number of the training example m

```python
m = x_train.shape[0]
print(f'The training examples is: {m}')
```

```python
# or another way to number of training examples
m = len(x_train)
print(f'The training examples is: {m}')
```

#### Training example x_i, y_i

```python
i = 0
x_i, y_i = x_train[i], y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Plot the data point
plt.scatter(x_train, y_train, c='r', marker='x');
plt.title('Housing Prices')
plt.ylabel('Prices (1000s of dollars)')
plt.xlabel('Size (1000s sft)')
plt.show()
```

#### Model Function

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros([m])
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
tmp_f_wb = compute_model_output(x_train, w, b)
```

```python
plt.scatter(x_train, tmp_f_wb, label='Our Pediction', c='b')
plt.scatter(x_train, y_train, marker='x', c = 'r', label='Actual Prediction')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
