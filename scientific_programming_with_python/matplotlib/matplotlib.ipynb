```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
%matplotlib inline
import numpy as np
x = np.linspace(0, 10, 100)
plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

```python
_, ax = plt.subplots()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x - 1), '-b')
ax.plot(x, np.sin(x - 2), color='g', linestyle='--')
ax.plot(x, np.sin(x - 3), color=(1.0, .2, .7), linestyle='-.')
ax.plot(x, np.sin(x - 5), color='0.75', linestyle=':')
ax.plot(x, np.sin(x - 6), color='chartreuse')
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-r', label='y')
plt.axis('equal')
plt.legend()
```
